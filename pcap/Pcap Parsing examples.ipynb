### Get list of domains from PCAP file

```python
from scapy.all import *
from scapy.layers.dns import DNSRR, DNS, DNSQR

domain_list= []
ip_list = []

# Read the PCAP file
packets = rdpcap('./pcap_example.pcapng')

# Iterate over the packets in the pcap file
for packet in packets:
    # get TCP/IP packet info
    if IP in packet:
        ip_src=packet[IP].src
        ip_dst=packet[IP].dst
    if TCP in packet:
        tcp_sport=packet[TCP].sport
        tcp_dport=packet[TCP].dport
    packet_info = "["+str(ip_src)+":"+str(tcp_sport)+"|"+str(ip_dst)  +":"+str(tcp_dport)+"]"
    ip_list.append(packet_info)
    
    # get DNS info from packet if applicable
    if packet.haslayer(DNSQR):
        query = packet[DNSQR].qname
        domain_list.append(query)

# deduplicate lists
dedup_domain_list = list(dict.fromkeys(domain_list))
dedup_ip_list = list(dict.fromkeys(ip_list))

print("\nList of Unique DNS queries in pcap:")
for el in dedup_domain_list:
    print(bytes.decode(el, 'utf-8'))
    
print("\nList of unique [IP Source:port|IP Destination:port] in pcap:")
for el in dedup_ip_list:
    print(el)
```
